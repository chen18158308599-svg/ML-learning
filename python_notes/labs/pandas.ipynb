```python
import numpy as np

np.random.seed(42)

a = np.random.randint(0,10)

np.random.randint(0,10)  # 模拟别的代码

b = np.random.randint(0,10)

print(a, b)
```

```python
rng = np.random.RandomState(42)

a = rng.randint(0,10)

np.random.randint(0,10)  # 别的代码

b = rng.randint(0,10)

print(a, b)
```
