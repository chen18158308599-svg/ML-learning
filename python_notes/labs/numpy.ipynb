# arrays
Unlike python lists, NumPy is constrained to arrays that contain the same type.

````python
np.array([1, 2, 3])
````

array automatically upcast objects to the same data type

````python
np.array([True,1,2,3])
````

## explanation: why numpy?
NumPy arrays are faster than Python lists mainly because:

- They store homogeneous data types

- Data are stored in contiguous memory

- They avoid Python object overhead

- Computation is performed in optimized C/Fortran code

````python
print(np.ones((3,4),dtype=str))
print(np.full((3,3),3.14))
print(np.arange(0,10,2))
print(np.linspace(0,10,5,dtype=int))# contains both start and end
print(np.random.random((3,3)))
np.empty(3)
````

``` np.empty() ``` allocates memory for an array without initializing its values. The elements therefore contain arbitrary data already present in memory. It is mainly used for performance when the array will be immediately filled with new values. It's faster than generating other fillers.

````python
print(np.full((3,4),fill_value=True,dtype=np.int_))
````
